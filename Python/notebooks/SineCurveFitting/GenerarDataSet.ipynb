```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# creando el conjunto de datos del seno
x_data = np.linspace(0, 3.2, num=1000)
pure_sin_y = np.sin(x_data)
```

```python
plt.plot(x_data, pure_sin_y, 'r.')
plt.grid(True)
plt.show()
```

```python
# generamos datos "ruidosos"
x_data = np.linspace(0, 3.2, num=1000)
y_data = np.sin(x_data) + 0.03 * np.random.normal(size=1000)
```

```python
plt.plot(x_data, y_data, 'b.')
plt.grid(True)
plt.show()
```

```python
import csv

with open('../../data/datasets/noisy_sin_singleCurve_dataset.csv', 'w', newline='') as fcsv:
    writer = csv.writer(fcsv)
    
    writer.writerow(['x', 'y'])
    
    for d in zip(x_data, y_data):
        writer.writerow(d)
```

```python
# tambien lo guardamos en formato npz
with open('../../data/datasets/noisy_sin_singleCurve_dataset.npz', 'wb') as f:
    np.save(f, x_data)
    np.save(f, y_data)
```
