```python
def _linear(z):
    return z

def _relu(z):
    return np.max([0, z])

def _dot(a, b):
    if len(a) != len(b):
        raise ValueError
    sum = 0;
    for i in range(len(a)):
        sum += a[i] * b[i]
        
    return sum

def _z_fun(x, w, b):
    return _dot(x, w) + b
```
