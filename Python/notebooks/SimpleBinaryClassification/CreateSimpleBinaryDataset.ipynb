```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns

# Large class Sep, Easy decision boundary
X,y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=2,class_sep=3,flip_y=0,weights=[0.5,0.5], random_state=17)
sns.scatterplot(x=X[:,0], y=X[:,1],hue=y)
plt.title("large")
plt.show();
```

```python
import pandas as pd

X = pd.DataFrame(data=X, columns=["f1", "f2"])
y = pd.Series(data=y)

data_df = pd.concat([X, y], axis=1)
data_df.rename(columns={0:'y'}, inplace=True)
```

```python
X_df = data_df[data_df['f1'] > 0.0]
X_df.loc[X_df["f2"] > 0.0, "y"] = 0
```

```python
sns.scatterplot(data=X_df, x=X_df['f1'], y=X_df['f2'], hue=X_df['y'])
plt.show()
```

```python
X_df = X_df[(X_df['f2'] > 0.5) | (X_df['f2'] < -0.5)]
```

```python
sns.scatterplot(data=X_df, x=X_df['f1'], y=X_df['f2'], hue=X_df['y'])
plt.show()
```

```python
X_df.to_csv('../data/datasets/simple_binary.csv', index=False, header=['Feature_1', 'Feature_2', 'Output'])
```
